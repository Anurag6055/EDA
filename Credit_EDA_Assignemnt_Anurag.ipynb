```python
#A function for finding the percentage of null values in each column in the given dataset
def dataset_null_values(df):
    return round((df.isnull().sum()*100/len(df)).sort_values(ascending = False),2)  

#Here we are rounding the number upto 2 decimal points and sorting so that 
#the columns with big percentage of missing values are shown at the top
```

```python
#Dividing the dataset based on the TARGET column
df1_0 = df1[df1.TARGET == 0]
df1_1 = df1[df1.TARGET == 1]
```

```python
#Defining a function so that we dont have to write all these again just 
#calling this function on the required columns will give us the required output

def plot_func(df,dim):
    fig = plt.figure(figsize=(13,10))
    
    #subplot 1 this subplot plots barplot of the column count
    ax1 = plt.subplot(221)
    df[dim].value_counts().plot.barh(ax = ax1)
    plt.title("Distribution of the attributes of the " + dim + " column")
    
    #subplot 2 this subplot plots that particular column with respec to the Target column
    ax2 = plt.subplot(222)
    sns.countplot(x=dim,hue='TARGET',data=df,ax=ax2)
    plt.xticks(rotation=45)
    plt.title("Each attribute count with respect to Target")
    plt.legend(['Repayer', 'Defaulter'])
    
    #subplot 3 this subplot plots the percentage of defaulter and repayer of the respective column
    ax3 = plt.subplot(223)
    df=pd.DataFrame()
    df['0'] = ((df1_0[dim].value_counts())/len(df1_0))
    df['1'] = ((df1_1[dim].value_counts())/len(df1_1))
    df.plot.bar(ax=ax3)
    plt.title("Percentage of Defaulter and Repayer in "+dim)
    plt.legend(['Repayer', 'Defaulter'])
    plt.xticks(rotation = 45)
    
    #fig.tight_layout()
    plt.show()
```

```python
def plot_func1(dataframe,dim):
    fig = plt.figure(figsize=(13,10))
    
    #subplot 1 
    ax1 = plt.subplot(221)
    dataframe[dim].value_counts().plot.barh(ax = ax1)
    plt.title("Distribution of the attributes of the " + dim + " column")
    
    #subplot 2
    ax2 = plt.subplot(222)
    sns.countplot(x=dim,hue='NAME_CONTRACT_STATUS',data=dataframe,ax=ax2)
    plt.xticks(rotation=45)
    plt.title("Each attribute count with respect to NAME_CONTRACT_STATUS")
    
    #subplot 3
    ax3 = plt.subplot(223)
    df=pd.DataFrame()
    df['Approved'] = ((complete_data_approved[dim].value_counts())/len(complete_data_approved))
    df['Canceled'] = ((complete_data_canceled[dim].value_counts())/len(complete_data_canceled))
    df['Refused'] = ((complete_data_refused[dim].value_counts())/len(complete_data_refused))
    df['Unused'] = ((complete_data_unused[dim].value_counts())/len(complete_data_unused))
    df.plot.bar(ax=ax3)
    plt.xticks(rotation=45)
    plt.title("Percentage for each")
    plt.show()
```

```python
#plot for gender from complete data

plot_func1(complete_data,"CODE_GENDER")
```

```python
# From what medium does the client know about us and the status of it

plot_func1(complete_data, "CHANNEL_TYPE")
```

```python
#plot for NAME_CLIENT_TYPE

plot_func1(complete_data, "NAME_CLIENT_TYPE")
```
